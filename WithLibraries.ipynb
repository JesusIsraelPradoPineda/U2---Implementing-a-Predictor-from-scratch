```python
path = "/content/Indicadores_municipales_sabana_DA (1).csv"
df = pd.read_csv(path, sep=",", encoding="ISO-8859-1")
```

```python
columns_to_drop = ['nom_mun']
df = df.drop(columns=columns_to_drop)
```

```python
columns_to_keep = [col for col in df.columns if all(x not in col for x in ['05', '10'])]

df = df[columns_to_keep]

df.to_csv('/content/Indicadores_municipales_sabana_DA (1).csv', index=False)
```

```python
df.head()
```

```python
df['gdo_rezsoc00'], _ = pd.factorize(df['gdo_rezsoc00'])

df.to_csv('/content/Indicadores_municipales_sabana_DA (1).csv', index=False)
```

```python
df.fillna(df.mean(), inplace=True)
df.to_csv('/content/Indicadores_municipales_sabana_DA (1).csv', index=False)
```

```python
train_size = int(0.8 * len(df))
train_df = df[:train_size]
test_df = df[train_size:]

X_train = train_df.drop('gini_00', axis=1)
y_train = train_df['gini_00']
X_test = test_df.drop('gini_00', axis=1)
y_test = test_df['gini_00']

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras import layers

model = tf.keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error')
```

```python
history = model.fit(
    X_train_scaled, y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.2
)
```

```python
loss = model.evaluate(X_test_scaled, y_test, verbose=0)
print(f"Test Loss: {loss}")
```

```python
import matplotlib.pyplot as plt

y_pred_train = model.predict(X_train_scaled).flatten()
y_pred_test = model.predict(X_test_scaled).flatten()

# Gráfica de las predicciones
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_train, y_pred_train, alpha=0.5)
plt.plot([0, 1], [0, 1], '--k')  # Se ajustan los límites de la gráfica para ir de 0 a 1
plt.xlabel('True GNI')
plt.ylabel('Predicted GNI')
plt.title('Training Set')
plt.xlim(0, 1)  # Fijar los límites del eje X
plt.ylim(0, 1)  # Fijar los límites del eje Y

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([0, 1], [0, 1], '--k')    # Se ajustan los límites de la gráfica para ir de 0 a 1
plt.xlabel('True GNI')
plt.ylabel('Predicted GNI')
plt.title('Test Set')
plt.xlim(0, 1)  # Fijar los límites del eje X
plt.ylim(0, 1)  # Fijar los límites del eje Y

plt.tight_layout()
plt.show()
```
