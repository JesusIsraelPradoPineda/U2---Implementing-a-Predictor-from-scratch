```python
train_size = int(0.8 * len(df))
train_df = df[:train_size]
test_df = df[train_size:]

X_train = train_df.drop('gini_00', axis=1)
y_train = train_df['gini_00']
X_test = test_df.drop('gini_00', axis=1)
y_test = test_df['gini_00']

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
def add_polynomial_features(X, degree):
    from itertools import combinations_with_replacement as combwr

    n_samples, n_features = X.shape
    comb = list(combwr(range(n_features), degree))

    X_poly = np.empty((n_samples, len(comb)))
    for i, c in enumerate(comb):
        X_poly[:, i] = np.prod(X[:, c], axis=1)

    return X_poly
degree = 2
X_train_poly = add_polynomial_features(X_train.values, degree)
X_test_poly = add_polynomial_features(X_test.values, degree)
```

```python
import numpy as np
def normalize(X):
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    X_norm = (X - mean) / std

    return X_norm, mean, std
X_train_poly_norm, mean, std = normalize(X_train_poly)
X_test_poly_norm = (X_test_poly - mean) / std
```

```python
def train_linear_regression_with_regularization(X, y, lambda_):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    identity_matrix = np.eye(X_b.shape[1])
    identity_matrix[0, 0] = 0

    theta = np.linalg.inv(X_b.T.dot(X_b) + lambda_ * identity_matrix).dot(X_b.T).dot(y)

    return theta
lambda_ = 1
theta_reg = train_linear_regression_with_regularization(X_train_poly_norm, y_train.values, lambda_)
```

```python
def mean_squared_error(y_true, y_pred):
    mse = np.mean((y_true - y_pred)**2)

    return mse
```

```python
def predict(X, theta):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    predictions = X_b.dot(theta)

    return predictions
```

```python
y_pred_train_reg = predict(X_train_poly_norm, theta_reg)
y_pred_test_reg = predict(X_test_poly_norm, theta_reg)

mse_train_reg = mean_squared_error(y_train.values, y_pred_train_reg)
mse_test_reg = mean_squared_error(y_test.values, y_pred_test_reg)

print("Train MSE with regularization:", mse_train_reg)
print("Test MSE with regularization:", mse_test_reg)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_train, y_pred_train_reg, alpha=0.5)
plt.plot([0, 1], [0, 1], '--k')
plt.xlabel('True GNI')
plt.ylabel('Predicted GNI')
plt.title('Training Set')

plt.subplot(1, 2, 2)
plt.scatter(y_test, y_pred_test_reg, alpha=0.5)
plt.plot([0, 1], [0, 1], '--k')
plt.xlabel('True GNI')
plt.ylabel('Predicted GNI')
plt.title('Test Set')

plt.tight_layout()
plt.show()
```
